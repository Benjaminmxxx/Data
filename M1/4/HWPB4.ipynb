A partir del CSV hospitales2.csv, generar un archivo CSV de salida, que contenga los siguientes campos en este orden: 

latitude

longitude

name

label

Con los correspondientes datos extraídos del CSV original, el campo name tiene que corresponder con la dirección del hospital, y el campo label con el nombre del hospital.

```python
df=pd.read_csv('./hospitales2.csv',sep=',')
```

```python
df.head(2)
```

```python
df['WKT']=df.WKT.str.replace('POINT','')
df['WKT']=df.WKT.str.replace('(','')
df['WKT']=df.WKT.str.replace(')','')
```

```python
df['WKT'][0:3]
```

```python
df.rename(columns={'WKT':'Cordenadas'},inplace=True)
```

```python
df.head(2)
```

```python
df[[' ','Latitud','Longitud']]=df['Cordenadas'].str.split(' ',expand=True)
```

```python
df.head()
```

```python
del df['Cordenadas']
del df[' ']
```

```python
dfPedido=df[['Latitud','Longitud','NOMBRE','DOM_NORMA']]
dfPedido.rename(columns={'NOMBRE':'label'},inplace=True)
dfPedido.rename(columns={'DOM_NORMA':'name'},inplace=True)
dfPedido
```

```python
dfPedido=dfPedido[['Latitud','Longitud','name','label']]
dfPedido
```
